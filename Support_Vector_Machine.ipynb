```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.datasets import make_blobs

# Generate synthetic linearly separable data
X, y = make_blobs(n_samples=100, centers=2, random_state=42, cluster_std=1.0)

# Visualize the generated data
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', s=30)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Linearly Separable Data")
plt.show()

# Train a Support Vector Classifier with a linear kernel
svc = SVC(kernel='linear', C=0.5)
svc.fit(X, y)

# Plot the decision boundary
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', s=30)

# Plot the decision boundary and margins
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# Create grid to evaluate model
xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = svc.decision_function(xy).reshape(XX.shape)

# Plot decision boundary and margins
ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.7,
           linestyles=['--', '-', '--'])

# Plot support vectors
ax.scatter(svc.support_vectors_[:, 0], svc.support_vectors_[:, 1], s=100,
           linewidth=1, facecolors='none', edgecolors='k')

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Support Vector Classifier with Linear Kernel")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.datasets import make_blobs

# Generate synthetic non-linearly separable data
X, y = make_blobs(n_samples=100, centers=2, random_state=42, cluster_std=2.5)

# Train a Support Vector Classifier with a polynomial kernel
svc_poly = SVC(kernel='poly', degree=3, C=0.7)
svc_poly.fit(X, y)

# Train a Support Vector Classifier with an RBF kernel
svc_rbf = SVC(kernel='rbf', C=0.7)
svc_rbf.fit(X, y)

# Plot decision boundary for polynomial kernel
plt.figure(figsize=(16, 6))
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', s=30)

ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z_poly = svc_poly.decision_function(xy).reshape(XX.shape)

ax.contour(XX, YY, Z_poly, colors='k', levels=[-1, 0, 1], alpha=0.7,
           linestyles=['--', '-', '--'])
ax.scatter(svc_poly.support_vectors_[:, 0], svc_poly.support_vectors_[:, 1], s=100,
           linewidth=1, facecolors='none', edgecolors='k')

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("SVC with Polynomial Kernel")

# Plot decision boundary for RBF kernel
plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', s=30)

ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z_rbf = svc_rbf.decision_function(xy).reshape(XX.shape)

ax.contour(XX, YY, Z_rbf, colors='k', levels=[-1, 0, 1], alpha=0.7,
           linestyles=['--', '-', '--'])
ax.scatter(svc_rbf.support_vectors_[:, 0], svc_rbf.support_vectors_[:, 1], s=100,
           linewidth=1, facecolors='none', edgecolors='k')

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("SVC with RBF Kernel")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.datasets import make_blobs

# Generate synthetic data with 3 classes
X, y = make_blobs(n_samples=300, centers=3, random_state=42, cluster_std=2.0)

# Plot the generated 3-class data
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', s=30)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Data with 3 Classes")
plt.show()

# Train a Support Vector Classifier with RBF kernel for multi-class classification
svc_rbf_multi = SVC(kernel='rbf', decision_function_shape='ovo')
svc_rbf_multi.fit(X, y)

# Plot decision boundaries for RBF kernel with 3 classes
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', s=30)

# Create grid to evaluate model
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

xx = np.linspace(xlim[0], xlim[1], 100)
yy = np.linspace(ylim[0], ylim[1], 100)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = svc_rbf_multi.decision_function(xy)

# Plot decision boundaries
for i in range(3):
    ax.contour(XX, YY, Z[:, i].reshape(XX.shape), levels=[0], alpha=0.7, colors='k')

# Plot support vectors
ax.scatter(svc_rbf_multi.support_vectors_[:, 0], svc_rbf_multi.support_vectors_[:, 1], s=100,
           linewidth=1, facecolors='none', edgecolors='k')

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("SVC with RBF Kernel for 3 Classes")
plt.show()
```
